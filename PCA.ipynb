```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
plt.close("all")
```

```python
dataset = pd.read_csv('pollution_us_2000_2016.csv',encoding = "ISO-8859-1")
```

```python
features=['State','NO2 Mean','NO2 AQI','O3 Mean','SO2 Mean','CO Mean']
```

```python
dataset = dataset[features]
dataset = dataset[dataset['State']=='New York']
dataset
```

```python
fig = plt.figure(figsize=(14, 7)) 
x = dataset['NO2 Mean']
plt.hist(x,bins=20,range=(x.min(), x.max()))
plt.xticks(fontsize=18) 
plt.yticks(fontsize=18)
plt.xlabel("NO2 Mean", fontsize=22) 
plt.show()
```

```python
fig, axs = plt.subplots(1,5,figsize=(14, 7))
axs[0].boxplot(np.array(dataset['NO2 Mean']))
axs[1].boxplot(np.array(dataset['NO2 AQI']))
axs[2].boxplot(np.array(dataset['O3 Mean']))
axs[3].boxplot(np.array(dataset['SO2 Mean']))
axs[4].boxplot(np.array(dataset['CO Mean']))
axs[0].set_title('NO2 Mean')
axs[1].set_title('NO2 AQI')
axs[2].set_title('O3 Mean')
axs[3].set_title('SO2 Mean')
axs[4].set_title('CO Mean')
plt.show()
```

```python
#empirical correlation coefficients between every pair of features
dataset.corr()
```

```python
#visualize 
corr = dataset.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```

```python
plt.scatter(dataset['NO2 Mean'],dataset['NO2 AQI'])
plt.xlabel("NO2 Mean", fontsize=22) 
plt.ylabel("NO2 AQI", fontsize=22) 
plt.show()
```

```python
#centering the dataset before PCA
dataset.mean()
centered_df=dataset.iloc[:,1:6]-dataset.mean()
centered_df.head()
np.array
```

```python
#Analysis the first two first principal directions 
pca = PCA()
pca.fit(centered_df)
#These numbers for each component are proportional to the Eigenvalues
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=2)
Y_sklearn = pca.fit_transform(centered_df)
Y_sklearn
```

```python
plt.scatter(Y_sklearn[:,0],Y_sklearn[:,1],c='red')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
